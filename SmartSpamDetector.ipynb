#READ DATA

```python
import pandas as pd
df=pd.read_csv("/content/spam.csv",encoding="ISO-8859-1")
df.head()
```

#REMOVE UNWANTED COLUMNS AND ROWS

```python
df.drop(["Unnamed: 2"],axis=1,inplace=True, errors='ignore')
df.drop(["Unnamed: 3"],axis=1,inplace=True, errors='ignore')
df.drop(["Unnamed: 4"],axis=1,inplace=True, errors='ignore')
```

```python
df
```

#CHANGE THE OUTPUT HAM & SPAM TO 1 & -1

```python
df['v1']=df['v1'].map({'ham':1,'spam':-1})
df.head()
```

#STORE V2 TO A VARIABLE CALLED SPAM

```python
spam=df.v2
spam
```

#NOW WE USE STEMMING AND TOKENIZATION METHOD

1.   STEMMING - CUTTING THE TAIL END OF THE WORD TO CONVERT IT TO ORGINALFORM
2.   TOKENIZATION - SPLITTING A WORD OR SENTENCE



```python
from nltk.stem import SnowballStemmer
from nltk import TweetTokenizer

ss=SnowballStemmer("english")
tk=TweetTokenizer()
```

#REJOIN IT

```python
spam=spam.apply(lambda x:" ".join([ss.stem(i.lower()) for i in tk.tokenize(x)]))
spam
```

#WE NEED TO CLEAN FREQUENT AND NO MEANING WORDS,SO WE USE STOPWARDS

*   STOPWARDS - TO REMOVE UNWANTED WORDS



```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
sw=stopwords.words("english")

sw
```

#RE-JOIN IT

```python
spam=spam.apply(lambda x:" ".join([i for i in x.split() if i not in sw]))
spam
```

#NOW WE ARE GOING TO FEATURE EXTRACTION-CONVERTION
#TFID VECTORIZER

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
x=tfidf.fit_transform(spam)
print(x)
print(tfidf.get_feature_names_out())
```

```python
df['v1'].head()
```

#NOW WE NEED TO SPLIT IT FOR TRAINING

```python
from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,df['v1'],test_size=.2,random_state=42)
```

#IMPORT AN ALGORITHM TO TRAIN THE MODEL

*   WE IMPORT RANDOMFOREST CLASSIFIER -(CLASSIFICATION ALGORITHM IN ML)



```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_predict=rf.predict(x_test)
y_predict
```

#FIND THE ACCURACY OF THE DATA




```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

print("CLASSIFICATION REPORT\n",classification_report(y_test,y_predict))
print("ACCURACY-SCORE :",accuracy_score(y_test,y_predict)*100)
```
